```python
import pandas as pd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#%matplotlib inline
#sns.set(color_codes=True)
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df.head()
df.tail(15)
df.dtypes
```
